```python
import io
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
fileName = "sreshta_temperature_test"

dateArray = list()
tempArray = list()


with open(fileName) as file:
    line = file.readline()
    while line:
        #remove the newline character at the end of line
        line = line.replace("\n", "")
        # split the datetime and temp fields
        fields = line.split(',')
        #d = "12-09-2020 06:32:44"
        #d = fields[0]
        #datetime_object = datetime.strptime(d, '%d-%m-%Y %H:%M:%S')
        #type(datetime_object)
        #print(datetime_object)
        dateStr = fields[0]
        tempStr = fields[1]
        # convert date from string format to date-object
        datetime_object = datetime.strptime(dateStr, '%d-%m-%Y %H:%M:%S')
        # append the date and temp to array
        dateArray.append(datetime_object)
        tempArray.append(float(tempStr))
        #print(line)
        # read next line and continue the loop until the end of the file is reached
        line = file.readline()
```

```python
plt.figure(figsize=(12,6.5))
plt.plot(dateArray, tempArray, color='blue', linestyle='dashed')
plt.xlabel('Datetime(mm-dd HH)')
plt.ylabel('Temperature')
plt.title('24 Hours Temp Observation at Srestha\'s House', fontsize=12)

plt.show()
```

```python
plt.figure(figsize=(6,4))
n, bins, patches = plt.hist(tempArray, 10, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Temperature')
plt.ylabel('Probability')
plt.title('Histogram of Temperature')
plt.title('Histogram - Temp Observation at Srestha\'s House', fontsize=12)


plt.grid(True)
plt.show()
```
